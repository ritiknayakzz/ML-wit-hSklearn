```python
## for in-built data structures
import numpy as np
 
## for plotting
import matplotlib.pyplot as plt
from matplotlib import style
 
style.use('seaborn-darkgrid')
font = {'family':'serif',
        'weight':'bold',
        'size':20}
plt.rc('font', **font)
plt.rcParams['figure.figsize'] = (15, 8)
 
## for ML
from sklearn import svm
from sklearn.datasets import make_blobs
```

```python
## Generate isotropic Gaussian blobs

X, y = make_blobs(n_samples = 40, centers = 2, random_state = 6)

print(X)
print(y)
print(len(X))
```

**Using Polynomial kernel**

```python
## Fit the model
## C is regularization parameter, must be positive

clf = svm.SVC(kernel = 'poly', C = 1000)
clf.fit(X, y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y, s = 30, cmap = 'jet')
```

```python
## plot the decision function

ax = plt.gca() ## get current axis
xlim = ax.get_xlim()
ylim = ax.get_ylim()

## create grid to evaluate model

xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)

YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T

## Z = clf.decision_function(xy).reshape(XX.shape)
Z = clf.predict(xy).reshape(XX.shape)

## plot decision boundary and margins

ax.contour(XX, YY, Z, colors = 'k', alpha = 1.0, levels = [-1, 0, 1],
           linestyles = ['--', '-', '--'])

## plot support vectors

clf_x, clf_y = clf.support_vectors_[:, 0], clf.support_vectors_[:, 1]

ax.scatter(X[:, 0], X[:, 1], c = y, s = 30, cmap = 'jet')
ax.scatter(clf_x, clf_y, s = 200,
           linewidth = 4, facecolors = 'none', edgecolors = 'k')

sv = ['support vector'] * 3
for num, name in enumerate(sv):
  plt.annotate(name, (clf_x[num], clf_y[num]),
               xytext = (clf_x[num] + 0.05, clf_y[num] - 1.0),
               arrowprops = dict(arrowstyle = '->',
                                 color = 'maroon'))
```
